```python
%matplotlib inline 
%pylab inline
import cv2
import numpy as np
np.set_printoptions(threshold=np.inf)
```

```python
img1 = cv2.cvtColor(cv2.imread('img/manyduckies.jpg'), cv2.COLOR_BGR2RGB)
img2 = cv2.cvtColor(cv2.imread('img/oneduck.jpg'), cv2.COLOR_BGR2RGB)
f, ax = plt.subplots(1, 2, figsize=(12,5))
ax[0].imshow(img1);
ax[1].imshow(img2);
```

```python
img_shape1 = img1.shape[0], img1.shape[1]
img_shape2 = img2.shape[0], img2.shape[1]
print(img_shape1, img_shape2)
```

```python
def rescale(x: float, L: float, U: float):
    return (x - L) / (U - L)
```

---------------------------------

```python
left_motor_img1 = np.zeros(shape=img_shape1, dtype="float32")
right_motor_img1 = np.zeros(shape=img_shape1, dtype="float32")
print(left_motor_img1.shape, right_motor_img1.shape)

left_motor_img2 = np.zeros(shape=img_shape2, dtype="float32")
right_motor_img2 = np.zeros(shape=img_shape2, dtype="float32")
```

```python
# -1 'inhibits' motor, +1 'stimulates' motor

left_motor_img1[:, :img_shape1[1]//2] = -1  
right_motor_img1[:, img_shape1[1]//2:] = -1  

left_motor_img2[:, :img_shape2[1]//2] = -1   
right_motor_img2[:, img_shape2[1]//2:] = -1  
```

```python
# print(left_motor_img1)
```

```python
lower_hsv = np.array([5, 100, 100])    
upper_hsv = np.array([30, 255, 255])

def preprocess(image: np.ndarray) -> np.ndarray:
    """ Returns a 2D array mask color segmentation of the image """
    hsv = cv2.cvtColor(image, cv2.COLOR_RGB2HSV)
    mask = cv2.inRange(hsv, lower_hsv, upper_hsv)//255
    return mask
```

```python
mask1 = preprocess(img1)
mask2 = preprocess(img2)
f, ax = plt.subplots(1, 2, figsize=(12,5))
ax[0].imshow(mask1)
ax[0].axvline(x=320)
ax[1].imshow(mask2)
ax[1].axvline(x=320)
```

```python
def create_side_constants(mask, left_motor, right_motor):
    # Mask is a matrix where there is 1 if found a duck (yellow), 0 if not a duck (not yellow)
    # Left motor is a matrix where 1 in left side and 0 if right side 
    # Right motor is a matrix where 1 in left side and 0 if right side 
        
    # Element-wise multiplication of mask * motor
    mask_vs_left_motor = mask * left_motor
    mask_vs_right_motor = mask * right_motor
    
    # DEBUG: Count occurance of a certain item
    unique, counts = numpy.unique(mask_vs_left_motor, return_counts=True)
    print(dict(zip(unique, counts)))

    # Element-wise multiplication
    L = float(np.sum(mask_vs_left_motor))
    R = float(np.sum(mask_vs_right_motor))
    
    print(L, R)
    return L, R
```

```python
# mask1[:,:] = 1
L1, R1 = create_side_constants(mask1, left_motor_img1, right_motor_img1)
L2, R2 = create_side_constants(mask2, left_motor_img2, right_motor_img2)
```

```python
print(img1.shape[0]*img1.shape[1])
```

```python
# After the sum, the constant L/R can be in the max img1.shape[0]*img1.shape[1]/2 (if all values are 1)
# To see, try uncomment the mask = 1
limit1 = img1.shape[0]*img1.shape[1]/2
L1_rescaled = rescale(L1, 0, limit1)
R1_rescaled = rescale(R1, 0, limit1)
print(L1_rescaled, R1_rescaled)

limit2 = img2.shape[0]*img2.shape[1]/2
L2_rescaled = rescale(L2, 0, limit2)
R2_rescaled = rescale(R2, 0, limit2)
print(L2_rescaled, R2_rescaled)
```

```python
# Tweak with the constants below to get to change velocity or stabilize movements
# Recall that pwm sets wheel torque, and is capped to be in [-1,1]
gain = 5.0
const = 0.2 # power under null activation - this ensures the robot does not halt

pwm_left1 = const + L1_rescaled * gain
pwm_right1 = const + R1_rescaled * gain
print(pwm_left1, pwm_right1)

pwm_left2 = const + L2_rescaled * gain
pwm_right2 = const + R2_rescaled * gain
print(pwm_left2, pwm_right2)
```

```python
left_top_motor_img1 = np.zeros(shape=img_shape1, dtype="float32")
left_bottom_motor_img1 = np.zeros(shape=img_shape1, dtype="float32")
right_top_motor_img1 = np.zeros(shape=img_shape1, dtype="float32")
right_bottom_motor_img1 = np.zeros(shape=img_shape1, dtype="float32")
print(left_top_motor_img1.shape, right_bottom_motor_img1.shape)

left_top_motor_img2 = np.zeros(shape=img_shape1, dtype="float32")
left_bottom_motor_img2 = np.zeros(shape=img_shape1, dtype="float32")
right_top_motor_img2 = np.zeros(shape=img_shape1, dtype="float32")
right_bottom_motor_img2 = np.zeros(shape=img_shape1, dtype="float32")
```

```python
left_top_motor_img1[:img_shape1[0]//2, :img_shape1[1]//2] = -1   
left_bottom_motor_img1[img_shape1[0]//2:, :img_shape1[1]//2] = -1
right_top_motor_img1[:img_shape1[0]//2, img_shape1[1]//2:] = -1 
right_bottom_motor_img1[img_shape1[0]//2:, img_shape1[1]//2:] = -1

left_top_motor_img2[:img_shape2[0]//2, :img_shape2[1]//2] = -1   
left_bottom_motor_img2[img_shape2[0]//2:, :img_shape2[1]//2] = -1
right_top_motor_img2[:img_shape2[0]//2, img_shape2[1]//2:] = -1 
right_bottom_motor_img2[img_shape2[0]//2:, img_shape2[1]//2:] = -1
```

```python
f, ax = plt.subplots(1, 4, figsize=(12,5))
ax[0].imshow(left_top_motor_img1, cmap='gray')
ax[0].axvline(x=left_top_motor_img1.shape[1]//2)
ax[1].imshow(left_bottom_motor_img1, cmap='gray')
ax[1].axvline(x=left_bottom_motor_img1.shape[1]//2)
ax[2].imshow(right_top_motor_img1, cmap='gray')
ax[2].axvline(x=right_top_motor_img1.shape[1]//2)
ax[3].imshow(right_bottom_motor_img1, cmap='gray')
ax[3].axvline(x=right_bottom_motor_img1.shape[1]//2)
```

```python
def create_side_constants_4directions(mask, left_top_motor, left_bottom_motor, right_top_motor, right_bottom_motor):
    # Mask is a matrix where there is 1 if found a duck (yellow), 0 if not a duck (not yellow)
    # Left top motor is a matrix where 1 in left top side and 0 the rest

    # Element-wise multiplication
    L_top = float(np.sum(mask * left_top_motor))
    L_bottom = float(np.sum(mask * left_bottom_motor))
    R_top = float(np.sum(mask * right_top_motor))
    R_bottom = float(np.sum(mask * right_bottom_motor))
    
    print(L_top, L_bottom, R_top, R_bottom)
    return L_top, L_bottom, R_top, R_bottom
```

```python
# mask1[:,:] = 1
L1_top, L1_bottom, R1_top, R1_bottom = create_side_constants_4directions(mask1, left_top_motor_img1, left_bottom_motor_img1, right_top_motor_img1, right_bottom_motor_img1)
L2_top, L2_bottom, R2_top, R2_bottom = create_side_constants_4directions(mask2, left_top_motor_img2, left_bottom_motor_img2, right_top_motor_img2, right_bottom_motor_img2)
```

```python
# After the sum, the constant L/R can be in the max img1.shape[0]*img1.shape[1]/2 (if all values are 1)
# To see, try uncomment the mask = 1
limit1 = img1.shape[0]*img1.shape[1]/2
L1_top_rescaled = rescale(L1_top, 0, limit1)
L1_bottom_rescaled = rescale(L1_bottom, 0, limit1)
R1_top_rescaled = rescale(R1_top, 0, limit1)
R1_bottom_rescaled = rescale(R1_bottom, 0, limit1)
print(L1_top_rescaled, L1_bottom_rescaled, R1_top_rescaled, R1_bottom_rescaled)

limit2 = img2.shape[0]*img2.shape[1]/2
L2_top_rescaled = rescale(L2_top, 0, limit2)
L2_bottom_rescaled = rescale(L2_bottom, 0, limit2)
R2_top_rescaled = rescale(R2_top, 0, limit2)
R2_bottom_rescaled = rescale(R2_bottom, 0, limit2)
print(L2_top_rescaled, L2_bottom_rescaled, R2_top_rescaled, R2_bottom_rescaled)
```

```python
# Tweak with the constants below to get to change velocity or stabilize movements
# Recall that pwm sets wheel torque, and is capped to be in [-1,1]
gain = 5.0
const = 0.2 # power under null activation - this ensures the robot does not halt

pwm_left1_4directions = const + L1_bottom_rescaled * gain + L1_top_rescaled
pwm_right1_4directions = const + (R1_top_rescaled * R1_bottom_rescaled) * gain
print(pwm_left1_4directions, pwm_right1_4directions)
```
